```python
# Crea script ottimizzato per ridurre l'uso della memoria
memory_optimized_patch = """
import torch
import numpy as np
import os
from sam2_instantmesh import instant_mesh_process, diffusion_image_generation
from omegaconf import OmegaConf

# Patch per ridurre l'uso della memoria
def patch_instant_mesh_process(images, debug_dir, name=None):
    device = torch.device('cuda')
    
    # Ottimizzazioni memoria
    torch.cuda.empty_cache()
    
    config_path = "./instantmesh/configs/instant-mesh-large.yaml"
    config = OmegaConf.load(config_path)
    IS_FLEXICUBES = True
    
    # PATCH 1: Ridurre la risoluzione per risparmiare memoria
    original_res = 320
    reduced_res = 256  # Riduci da 320 a 256
    
    # PATCH 2: Usare half precision per ridurre memoria
    torch.backends.cudnn.allow_tf32 = True
    torch.backends.cuda.matmul.allow_tf32 = True
    
    from instantmesh.src.utils.train_util import instantiate_from_config
    import torchvision
    from instantmesh.src.utils.camera_util import get_zero123plus_input_cameras
    from sam2_instantmesh import save_obj
    
    instant_mesh = instantiate_from_config(config.model_config)
    model_ckpt_path = config.infer_config.model_path
    state_dict = torch.load(model_ckpt_path, weights_only=True)['state_dict']
    state_dict = {k[14:]: v for k, v in state_dict.items() if k.startswith('lrm_generator.')}
    instant_mesh.load_state_dict(state_dict, strict=True)
    
    instant_mesh = instant_mesh.to(device)
    # PATCH 3: Ridurre la risoluzione della geometria
    instant_mesh.init_flexicubes_geometry(device, fovy=30.0)
    instant_mesh = instant_mesh.eval()
    
    # PATCH 4: Ridurre la risoluzione delle immagini
    images = images.unsqueeze(0).to(device)
    images = torchvision.transforms.functional.resize(images, reduced_res, interpolation=3, antialias=True).clamp(0, 1)
    
    input_cameras = get_zero123plus_input_cameras(batch_size=1, radius=4.0 * 1.0).to(device)
    
    with torch.no_grad():
        # PATCH 5: Pulire la memoria prima dell'operazione critica
        torch.cuda.empty_cache()
        
        planes = instant_mesh.forward_planes(images, input_cameras)
        
        # PATCH 6: Modificare i parametri di inferenza per ridurre la memoria
        infer_config = config.infer_config.copy()
        if hasattr(infer_config, 'chunk_size'):
            infer_config.chunk_size = min(infer_config.get('chunk_size', 5), 3)
        if hasattr(infer_config, 'render_resolution'):
            infer_config.render_resolution = min(infer_config.get('render_resolution', 512), 256)
            
        mesh_path_idx = os.path.join(debug_dir, f'mesh_{name}.obj')
        
        # PATCH 7: Estrarre la mesh con parametri ridotti
        mesh_out = instant_mesh.extract_mesh(planes, use_texture_map=False, **infer_config)
        vertices, faces, vertex_colors = mesh_out
        save_obj(vertices, faces, vertex_colors, mesh_path_idx)
        print(f"Mesh saved to {mesh_path_idx}")
        
    del instant_mesh, planes, images
    torch.cuda.empty_cache()

# Applica le patch
import sam2_instantmesh
sam2_instantmesh.instant_mesh_process = patch_instant_mesh_process

print("Patch di ottimizzazione memoria applicata!")
"""

# Salva il patch
with open('memory_patch.py', 'w') as f:
    f.write(memory_optimized_patch)

# Esegui il patch nell'environment conda any6d
!conda run -n any6d python memory_patch.py
```

```python
# Modifica la configurazione di InstantMesh per ridurre l'uso della memoria
config_patch = """
import yaml
from omegaconf import OmegaConf

# Carica la configurazione esistente
config_path = './instantmesh/configs/instant-mesh-large.yaml'
config = OmegaConf.load(config_path)

# Modifica i parametri per ridurre l'uso della memoria
if 'infer_config' in config:
    # Riduce la risoluzione di rendering
    if 'render_resolution' in config.infer_config:
        config.infer_config.render_resolution = 256
    else:
        config.infer_config.render_resolution = 256
    
    # Riduce il chunk size se presente
    if 'chunk_size' in config.infer_config:
        config.infer_config.chunk_size = 2
    else:
        config.infer_config.chunk_size = 2
        
    # Aggiungi parametri di ottimizzazione memoria
    config.infer_config.low_memory_mode = True

# Salva la configurazione modificata
with open('./instantmesh/configs/instant-mesh-large-lowmem.yaml', 'w') as f:
    OmegaConf.save(config, f)

print("Configurazione ottimizzata salvata in instant-mesh-large-lowmem.yaml")
"""

with open('config_patch.py', 'w') as f:
    f.write(config_patch)

!conda run -n any6d python config_patch.py
```

```python
# Se il fix precedente non funziona, crea un nuovo run_demo.py
new_run_demo = '''
import argparse
import copy
import os
import numpy as np
import trimesh
import torch
import gc

# Ottimizzazioni memoria
os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'expandable_segments:True'
torch.backends.cudnn.allow_tf32 = True
torch.backends.cuda.matmul.allow_tf32 = True

from estimater import Any6D
from sam2_instantmesh import preprocess_image, get_bounding_box, running_sam_box, diffusion_image_generation, instant_mesh_process

def align_mesh_to_coordinate(mesh):
    centroid = mesh.vertices.mean(axis=0)
    mesh.vertices -= centroid
    return mesh

if __name__ == "__main__":
    parser = argparse.ArgumentParser()
    parser.add_argument('--ycb_model_path', type=str, default='./bop_toolkit/datasets')
    parser.add_argument('--img_to_3d', action='store_true')
    args = parser.parse_args()

    obj = 'mustard'
    color_path = './demo_data/color.png'
    depth_path = './demo_data/depth.png'
    camera_path = './demo_data/836212060125_640x480.yml'
    label_path = './demo_data/labels.npz'
    mesh_path = './demo_data/mustard.obj'
    
    save_path = './demo_data'
    img_to_3d = args.img_to_3d

    color = np.array(Image.open(color_path))
    depth = np.load(depth_path)
    label = np.load(label_path, allow_pickle=True)

    obj_list = ['002_master_chef_can', '003_cracker_box', '004_sugar_box', '005_tomato_soup_can', '006_mustard_bottle']
    obj_idx = {'002_master_chef_can': 1, '003_cracker_box': 2, '004_sugar_box': 3, '005_tomato_soup_can': 4, '006_mustard_bottle': 5}

    if obj == 'mustard':
        obj_num = obj_idx['006_mustard_bottle']
    else:
        obj_num = obj_idx[obj]

    mask = np.where(label['seg'] == obj_num, 255, 0).astype(np.bool_)

    if img_to_3d:
        cmin, rmin, cmax, rmax = get_bounding_box(mask).astype(np.int32)
        input_box = np.array([cmin, rmin, cmax, rmax])[None, :]
        mask_refine = running_sam_box(color, input_box)

        input_image = preprocess_image(color, mask_refine, save_path, obj)
        images = diffusion_image_generation(save_path, save_path, obj, input_image=input_image)
        
        # Pulisci memoria prima di InstantMesh
        torch.cuda.empty_cache()
        gc.collect()

        # Applica patch di memoria
        exec(open('memory_patch.py').read())

        instant_mesh_process(images, save_path, obj)

        # Pulisci memoria dopo InstantMesh
        torch.cuda.empty_cache()
        gc.collect()

        mesh = trimesh.load(os.path.join(save_path, f'mesh_{obj}.obj'))
        mesh = align_mesh_to_coordinate(mesh)
        mesh.export(os.path.join(save_path, f'center_mesh_{obj}.obj'))

        mesh = trimesh.load(os.path.join(save_path, f'center_mesh_{obj}.obj'))
    else:
        mesh = trimesh.load(mesh_path)

    est = Any6D(symmetry_tfs=None, mesh=mesh, debug_dir=save_path, debug=2)

    from foundationpose.Utils import *
    to_origin, extents = trimesh.bounds.oriented_bounds(mesh)
    bbox = np.stack([-extents/2, extents/2], axis=0).reshape(2, 3)

    scorer = est.score_object(color, depth, camera_path, bbox, [0, 0, 0, 1, 0, 0, 0], mask)
    print('Score:', scorer)
'''

# Sostituisci il file con la versione corretta
with open('run_demo_new.py', 'w') as f:
    f.write(new_run_demo)

print("Creato run_demo_new.py con ottimizzazioni integrate")

# Sostituisci il file originale
import shutil
shutil.copy('run_demo_new.py', 'run_demo.py')
print("File run_demo.py sostituito con la versione corretta")
```
